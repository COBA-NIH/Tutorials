```python
import seaborn as sns
import pandas
import matplotlib.pyplot as plt
```

```python
t = range(150)
p_80 = [0.8**x for x in t]
p_90 = [0.9**x for x in t]
p_95 = [0.95**x for x in t]
p_99 = [0.99**x for x in t]

df = pandas.DataFrame({"Accuracy 0.8":p_80, "Accuracy 0.9":p_90, "Accuracy 0.95":p_95, "Accuracy 0.99":p_99, })
```

```python
sns.lineplot(df)
plt.xlabel("Number of timepoints")
plt.ylabel("Fraction of tracks accurately called")
```
